```
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils
import numpy as np
import tensorflow as tf
```

```
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```
train_len,w,l=x_train.shape
inp_train = x_train.reshape(train_len,w*l)/255
inp_train.shape
```

```
out_train = utils.to_categorical(y_train,10)
out_train.shape
```

```
model = Sequential()
model.add(Dense(800,input_shape=(w*l,),activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```
model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])
model.summary()
```

```
model.fit(inp_train,out_train,batch_size=200,epochs=100,validation_split=0.2)
```

```
inp_test = x_test.reshape(10000,28*28)/255
inp_test.shape
```

```
out_test = utils.to_categorical(y_test,10)
out_test.shape
```

```
model.evaluate(inp_test,out_test)
```
